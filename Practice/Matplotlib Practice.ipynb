```python
x = np.linspace(1, 21)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))

axes[0][0].plot(x, x**2)
axes[0][1].plot(x, x)
axes[1][0].plot(x, np.log(x))
axes[1][1].plot(x, x**5)

plt.show()
```

```python
x = np.arange(1,30)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))

axes[0][0].plot(x, np.sin(x))
axes[0][1].plot(x, x+x**2)
axes[1][0].plot(x, np.log(x))
axes[1][1].plot(x, 5**x)

plt.show()
```

```python
x = np.linspace(0, 5, 11)

fig, axes = plt.subplots(nrows=1, ncols=3)
axes[0].plot(x, x**2, x, x**4)
axes[0].set_title('default axis range')
axes[1].plot(x, x**2, x, x**4)
axes[1].axis('tight')
axes[1].set_title('tight axis range')
axes[2].plot(x, x**2, x, x**4)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range");

plt.show()
```

```python
x = np.linspace(0, 5, 20)
fig, axes = plt.subplots(nrows=1, ncols=3)
axes[0].plot(x, x**x, x, x/8)
axes[0].set_title('default axis range')
axes[1].plot(x, x**x, x, x/8)
axes[1].axis('tight')
axes[1].set_title('tight axis range')
axes[2].plot(x, x**x, x, x/8)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range");

plt.show()
```

```python
sat_scores_df = pd.read_csv(filepath_or_buffer='2012-sat-results.csv')
sat_scores_df.head()
```

```python
sat_scores_df=sat_scores_df[sat_scores_df["Num of SAT Test Takers"] != "s"]
sat_scores_df['Num of SAT Test Takers'] = pd.to_numeric(sat_scores_df['Num of SAT Test Takers'])
sat_scores_df['SAT Critical Reading Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Critical Reading Avg. Score'])
sat_scores_df['SAT Math Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Math Avg. Score'])
sat_scores_df['SAT Writing Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Writing Avg. Score'])
```

```python
sat_scores_df['Avg. Total SAT Score'] = sat_scores_df['SAT Critical Reading Avg. Score'] + sat_scores_df['SAT Math Avg. Score'] + sat_scores_df['SAT Writing Avg. Score']
sat_scores_df.head()
```

```python
sat_scores_df.info()
```

```python
top_5 = sat_scores_df.sort_values(by=['Num of SAT Test Takers'], ascending=False).head(5)

colors = ['blue', 'purple', 'lightskyblue', 'violet','navy']
patches, texts = plt.pie(top_5['Num of SAT Test Takers'], colors=colors, shadow=True, startangle=90)
plt.legend(patches, top_5['SCHOOL NAME'])

plt.show()
```

```python
top_5 = sat_scores_df.sort_values(by=['Avg. Total SAT Score'], ascending=False).head(5)

plt.bar(top_5['SCHOOL NAME'], top_5['Avg. Total SAT Score'])
plt.title('Top 5 SAT Scoring Schools')
plt.show()
```
