```python
sat_scores_df = pd.read_csv(filepath_or_buffer='2012-sat-results.csv')
```

```python
sat_scores_df.head(30)
```

```python
sat_scores_df.info()
```

```python
sat_scores_df=sat_scores_df[sat_scores_df["Num of SAT Test Takers"] != "s"]
```

```python
sat_scores_df.head(30)
```

```python
sat_scores_df['Num of SAT Test Takers'] = pd.to_numeric(sat_scores_df['Num of SAT Test Takers'])
sat_scores_df['SAT Critical Reading Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Critical Reading Avg. Score'])
sat_scores_df['SAT Math Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Math Avg. Score'])
sat_scores_df['SAT Writing Avg. Score'] = pd.to_numeric(sat_scores_df['SAT Writing Avg. Score'])
```

```python
sat_scores_df.info()
```

```python
sat_scores_df['Avg. Total SAT Score'] = sat_scores_df['SAT Critical Reading Avg. Score'] + sat_scores_df['SAT Math Avg. Score'] + sat_scores_df['SAT Writing Avg. Score']
```
